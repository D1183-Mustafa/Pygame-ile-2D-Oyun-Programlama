```python
import pygame
pygame.init()

en = 800
boy = 600

ekran = pygame.display.set_mode((en, boy))
pygame.display.set_caption = "uzay oyunu"

arkaplan = pygame.image.load("arka plan.jpg")

class oyuncu():
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def çiz(self):
        oyuncuresim = pygame.image.load("pngwing.com.png")
        ekran.blit(oyuncuresim, (self.x, self.y))
        
class düşman():
    def __init__(self, x , y):
        self.x = x
        self.y = y
        self.puan = 0
        
    def çiz(self):
        düşmanresim = pygame.image.load("uzaylı.png")
        ekran.blit(düşmanresim,(self.x, self.y))
        self.y += 0.2
        
    def vurulma(self):
        for mermi in mermiler:
            if mermi.x > self.x and mermi.x < self.x + 50 and mermi.y > self.y and mermi.y < self.x + 50:
                düşmanlar.remove(self)
                mermiler.remove(mermi)
                self.puan += 10
        
        
class mermi():
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def çiz(self):
        pygame.draw.rect(ekran,(255,0,0), pygame.Rect(self.x,self.y,2,5))
        self.y -= 2
    
def mesaj(mesaj):
    font = pygame.font.SysFont("", 50)
    msj = font.render(mesaj,True,(255,255,255))
    yazıboyut = msj.get_size()
    ekran.blit(msj,(en // 2- yazıboyut[0] // 2, 160))
        
oyuncu = oyuncu(en / 2, boy - 100)
düşmanlar = []
mermiler = []
for x in range(1,11):
    for y in range(1,4):
        düşmanlar.append(düşman(x * 50, y * 50))
    
başla = True
fps = pygame.time.Clock()
while başla:
    fps.tick(60)
    ekran.blit(arkaplan,(0,0))
    oyuncu.çiz()
    
    for düşman in düşmanlar:
        düşman.çiz()
        düşman.vurulma()
        if düşman.y > boy - 80:
            mesaj("Kaybettin!!!")
            
    if len(düşmanlar) <= 0:
        mesaj("Kazandınız !!!")
        
    for mermi in mermiler:
        mermi.çiz()
    
    for i in pygame.event.get():
        if i.type == pygame.QUIT:
            başla = False
        if i.type == pygame.KEYDOWN and i.key == pygame.K_UP:
            mermiler.append(mermi(oyuncu.x + 32, oyuncu.y))
            
    tuş = pygame.key.get_pressed()
    
    if tuş[pygame.K_LEFT]:
        if oyuncu.x > 20:
            oyuncu.x -= 5
    if tuş[pygame.K_RIGHT]:
        if oyuncu.x < en - 20 - 50:
            oyuncu.x += 5
            
            
            
    pygame.display.update()
```
