```python
import pygame

pygame.init()  # pygame kütüphanesini başlat

ekran = pygame.display.set_mode((800,600))  # 800 x 600 ölçüsünde ekran oluşturur.
pygame.display.set_caption("Oyun denemesi")

font = pygame.font.SysFont("Helvetica",18)
yazı = font.render("Oyunuma Hoş Geldiniz!", True,(255,0,0))
yazıboyutu = yazı.get_size()

başla = True
clock = pygame.time.Clock()
while başla:
    clock.tick(60)
    for i in pygame.event.get():
        if i.type == pygame.QUIT:
            başla = False
        print(i)
        
        ekran.fill((0,0,0))
        ekran.blit(yazı,(400 - yazıboyutu[0] / 2,20))
        
        pygame.draw.rect(ekran,(250,0,0), (60,60,100,100))
        pygame.draw.circle(ekran,(0,255,0), (60,160), 30)
        pygame.draw.line(ekran,(0,0,255), (60,130),(250,250),6)
        
        pygame.display.flip()
```
